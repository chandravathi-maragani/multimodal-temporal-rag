```python
import json
import re
from pathlib import Path
```

```python
#to extract the dates from our notebook json files but it failed as 'date' key is nested inside and present in 'pages'
def inspect_raw_backup_dates(json_dir_path: str):
    """
    Scans the JSON backup directory and prints every unique raw value 
    found inside the 'date' key to inspect the exact string format.
    """
    raw_dates_found = set()
    json_path = Path(json_dir_path)
    
    if not json_path.exists():
        print(f"Error: The directory '{json_dir_path}' was not found.")
        return
        
    for file_path in json_path.glob("*.json"):
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)
                
            # Grab exactly what is in the 'date' key right now
            date_val = data.get("date")
            if date_val is not None:
                raw_dates_found.add(str(date_val).strip())
                
        except Exception as e:
            print(f"Could not read file {file_path.name}: {e}")
            continue

    print("\n=== RAW DATE VALUES FOUND IN YOUR JSON FILES ===")
    if not raw_dates_found:
        print("No 'date' keys found or all values were empty.")
    else:
        for d in sorted(list(raw_dates_found)):
            print(f"- '{d}'")
    print("================================================\n")

# To run this, replace the string below with your actual folder path:
# inspect_raw_backup_dates("path/to/your/json_backups")

inspect_raw_backup_dates(r"D:\Multimodal-Temporal-RAG\my_notebook_json_backups")
```

```python
#to extract the json file structure to make sure where exactly the 'date' key is present
def peek_at_first_json_structure(json_dir_path: str):
    json_path = Path(json_dir_path)
    # Grab just the very first JSON file in the directory
    first_file = next(json_path.glob("*.json"), None)
    
    if not first_file:
        print(f"No JSON files found in {json_dir_path}")
        return
        
    with open(first_file, "r", encoding="utf-8") as f:
        data = json.load(f)
        
    print(f"=== STRUCTURE OF FILE: {first_file.name} ===")
    print("Available top-level keys:", list(data.keys()))
    print("\nFull file content snippet:")
    print(json.dumps(data, indent=2)[:1000]) # Prints the first 1000 characters
    print("==========================================")

# Run it against your backup directory:
peek_at_first_json_structure(r"D:\Multimodal-Temporal-RAG\my_notebook_json_backups")
```

```python
#the correct revised function to extract the dates
def inspect_nested_backup_dates(json_dir_path: str):
    """
    Scans the JSON backup directory, navigates into the 'pages' array,
    and collects every unique raw string found inside the nested 'date' key.
    """
    raw_dates_found = set()
    json_path = Path(json_dir_path)
    
    if not json_path.exists():
        print(f"Error: The directory '{json_dir_path}' was not found.")
        return
        
    for file_path in json_path.glob("*.json"):
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)
                
            # Navigate into the nested 'pages' list
            pages = data.get("pages", [])
            for page in pages:
                date_val = page.get("date")
                if date_val is not None:
                    raw_dates_found.add(str(date_val).strip())
                
        except Exception as e:
            print(f"Could not read file {file_path.name}: {e}")
            continue

    print("\n=== RAW NESTED DATE VALUES FOUND ===")
    if not raw_dates_found:
        print("No 'date' keys found inside the pages arrays.")
    else:
        for d in sorted(list(raw_dates_found)):
            print(f"- '{d}'")
    print("====================================\n")

# Run this inside a cell using your path:

inspect_nested_backup_dates(r"D:\Multimodal-Temporal-RAG\my_notebook_json_backups")
```

```python
#function to extract all the filenames, topics and dates(converted to year-date format without changing the files)
def generate_metadata_source_of_truth(json_dir_path: str) -> dict:
    """
    Scans the JSON backup directory, navigates into the nested 'pages' array,
    and builds strict, unique lists of filenames, topics, and transformed strings
    for the fuzzy search matching layer.
    """
    unique_filenames = set()
    unique_topics = set()
    unique_dates = set()

    json_path = Path(json_dir_path)
    if not json_path.exists():
        print(f"Warning: Directory {json_dir_path} does not exist.")
        return {"filenames":[], "topics":[], "dates":[]}

    #Month mapping helper to identify 2025 vs 2026 boundaries
    months_2025 = {"sep", "oct", "nov", "dec", "september", "october", "november", "december"}

    for file_path in json_path.glob("*.json"):
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)
            
            # 1.Capture and format top level Filename
            fallback_filename = file_path.stem + ".pdf"
            filename_val = data.get("filename", fallback_filename)

            if filename_val:
                filename_val = filename_val.strip()
                
                if filename_val.lower().endswith('.json'):
                    filename_val = filename_val[:-5]
                if not filename_val.lower().endswith('.pdf'):
                    filename_val += ".pdf"

                filename_val = filename_val[0].upper()+filename_val[1:]
                unique_filenames.add(str(filename_val))

            # 2. Iterate through the nested pages array for Topic and Dates
            pages = data.get("pages", [])
            for page in pages:
                # Process Topic
                topic_val = page.get("topic")
                if topic_val:
                    unique_topics.add(str(topic_val).lower().strip())
                
                # Process Date with localized year injection rules
                raw_date = page.get("date")
                if raw_date and str(raw_date).strip().lower() not in ["unknown", "n/a", "undated"]:
                    clean_date = str(raw_date).strip()
                    lower_date = clean_date.lower()
                    
                    # Guard: If a 4-digit year is already baked in, keep it as-is
                    if re.search(r'\b(2025|2026)\b', clean_date):
                        unique_dates.add(clean_date)
                    else:
                        # Inspect the month token to determine the year split
                        # e.g., 'Apr-16' -> '2026 Apr-16', 'Dec 15' -> '2025 Dec 15'
                        if any(m in lower_date for m in months_2025):
                            unique_dates.add(f"2025 {clean_date}")
                        else:
                            unique_dates.add(f"2026 {clean_date}")
                            
        except Exception as e:
            print(f"Skipping file {file_path.name} due to parsing error: {str(e)}")
            continue

    metadata_directory = {
        "filenames" : sorted(list(unique_filenames)),
        "topics" : sorted(list(unique_topics)),
        "dates" : sorted(list(unique_dates))
    }
    print(f"-> Extracted {len(metadata_directory['filenames'])} Files")
    print(f"-> Extracted {len(metadata_directory['topics'])} Topics")
    print(f"-> Extracted {len(metadata_directory['dates'])} Dates")
    return metadata_directory
        

generate_metadata_source_of_truth(r"D:\Multimodal-Temporal-RAG\my_notebook_json_backups")     
        
```

```python
import difflib
import re

def extract_metadata_via_fuzzy_match(user_query: str, metadata_directory: dict) -> dict:
    """
    Scans a natural language user query and searches for the closest matches
    against the known lists of filenames, topics, and dates using fuzzy matching.
    
    Returns a dictionary of extracted filters: {'filename': ..., 'topic': ..., 'date': ...}
    If no confident match is found for a field, it returns None for that key.
    """
    query_clean = user_query.strip().lower()
    
    extracted_filters = {
        "filename": None,
        "topic": None,
        "date": None
    }
    
    # ----------------------------------------------------
    # 1. Fuzzy Match Filename
    # ----------------------------------------------------
    # Look for matching base names or parts (e.g., "agentic ai" matching "Agentic_AI.pdf")
    best_file_match = None
    highest_file_ratio = 0.0
    
    for filename in metadata_directory.get("filenames", []):
        # Strip extension and replace underscores for better sentence matching
        base_name = filename.lower().replace(".pdf", "").replace("_", " ")
        
        # Check if the filename base is directly mentioned in the query
        if base_name in query_clean or filename.lower() in query_clean:
            extracted_filters["filename"] = filename
            break
        
        # Fallback to a sequence matcher ratio if it's a close spelling match
        ratio = difflib.SequenceMatcher(None, base_name, query_clean).ratio()
        if ratio > highest_file_ratio and ratio > 0.4:  # Adjust threshold as needed
            highest_file_ratio = ratio
            best_file_match = filename
            
    if not extracted_filters["filename"] and highest_file_ratio > 0.6:
        extracted_filters["filename"] = best_file_match

    # ----------------------------------------------------
    # 2. Fuzzy Match Topic
    # ----------------------------------------------------
    # Since topics are already lowercase strings, we can check for direct containment or close matches
    best_topic_match = None
    highest_topic_ratio = 0.0
    
    for topic in metadata_directory.get("topics", []):
        if topic in query_clean:
            extracted_filters["topic"] = topic
            break
            
        ratio = difflib.SequenceMatcher(None, topic, query_clean).ratio()
        if ratio > highest_topic_ratio and ratio > 0.5:
            highest_topic_ratio = ratio
            best_topic_match = topic
            
    if not extracted_filters["topic"] and highest_topic_ratio > 0.7:
        extracted_filters["topic"] = best_topic_match

    # ----------------------------------------------------
    # 3. Match Date (With Token-Based Fallback)
    # ----------------------------------------------------
    for date_str in metadata_directory.get("dates", []):
       # 1. Clean up the source date tokens
       # e.g., "2025 Dec 10" -> {"dec", "10"} or "2026 02-Jan" -> {"02", "jan"}
       raw_date_part = date_str.replace("2025 ", "").replace("2026 ", "").lower()
       source_tokens = set(re.findall(r'[a-z0-9]+', raw_date_part))
    
       # Clean up the user query tokens (stripping 'st', 'nd', 'rd', 'th' from numbers)
       # e.g., "10th October" -> {"10", "october"}
       clean_query_processed = re.sub(r'\b(\d+)(st|nd|rd|th)\b', r'\1', query_clean)
       query_tokens = set(re.findall(r'[a-z0-9]+', clean_query_processed))

       # 2. Direct containment check first (fast pass)
       if raw_date_part in query_clean or date_str.lower() in query_clean:
          extracted_filters["date"] = date_str
          break

       # 3. Token Intersection Check (handles "10 December", "10th dec", etc.)
       # If all tokens of the source date (like 'dec' and '10') are present in the user query tokens
       if source_tokens.issubset(query_tokens):
          extracted_filters["date"] = date_str
          break
        
       # 4. Handle Month Abbreviations (e.g., source has 'december' but user typed 'dec')
       # Check if the numeric day matches, and the month token starts with the user's abbreviation
       day_token = next((t for t in source_tokens if t.isdigit()), None)
       month_token = next((t for t in source_tokens if not t.isdigit()), None)
    
       if day_token in query_tokens and month_token:
          # Check if any token in the query starts with our month prefix (min 3 chars, e.g., "dec" in "december")
          if any(q_tok.startswith(month_token[:3]) or month_token.startswith(q_tok[:3]) for q_tok in query_tokens if not q_tok.isdigit()):
              extracted_filters["date"] = date_str
              break

    return extracted_filters
```
